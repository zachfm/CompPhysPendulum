```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Stores data regarding the current and past states of a single pendulum
class Pendulum:
    def __init__(self, theta0, vel0):
        # Angular velocity
        self.vel = [vel0]
        # Angular position
        self.theta = [theta0]
        # Vector of timesteps
        self.t = [0,dt]
        # Initialize velocity and position with the Euler integration 
        # scheme
        accel = calAccel(self, theta0);
        self.vel.append(self.vel[-1]+accel*dt);
        self.theta.append(self.theta[-1]+self.vel[-1]*dt);
```

```python

# Purpose: Calculate the acceleration of a driven pendulum, given its
# current angular position and the current simulation time
# Arguments: A Pendulum object and the Pendulum's current angle
# Retuns: The acceleration of the pendulum at this point in time
def calAccel(pend, theta):
    accel = -(k**2)*math.sin(theta)+ \
              gamma*(k**2)*math.sin(pend.t[-1]*omega)
    return accel

def calAccelUpper(pend, theta):
    accel = -(k**2)*math.sin(theta)+ \
              gamma*(k**2)*math.sin(pend.t[-1]*omega)
    return accel

def calAccelLower(pend, theta):
    accel = -(k**2)*math.sin(theta)+ \
              gamma*(k**2)*math.sin(pend.t[-1]*omega)
    return accel

# Purpose: Perform a single step in a Bashforth-Adams integration of a
# driven simple pendulum
# Arguments: The Pendulum to integrate over
# Returns: The same Pendulum, one integration step further on
def bashforthAdamsUpper(pendUpper, pendLower):
    vn = pendUpper.vel[-1] + 1.5*dt*calAccelUpper(pendUpper, pendUpper.theta[-1]) - \
         0.5*dt*calAccelUpper(pendUpper, pendUpper.theta[-2])
    thn = pendUpper.theta[-1] + 1.5*dt*pendUpper.vel[-1] - 0.5*dt*pendUpper.vel[-2]
    pendUpper.vel.append(vn)
    pendUpper.theta.append(thn)
    pendUpper.t.append(pendUpper.t[-1] + dt)
    return pendUpper

def bashforthAdamsLower(pendUpper, pendLower):
    vn = pendLower.vel[-1] + 1.5*dt*calAccelLower(pendLower, pendLower.theta[-1]) - \
         0.5*dt*calAccelLower(pendLower, pendLower.theta[-2])
    thn = pendLower.theta[-1] + 1.5*dt*pendLower.vel[-1] - 0.5*dt*pendLower.vel[-2]
    pendLower.vel.append(vn)
    pendLower.theta.append(thn)
    pendLower.t.append(pendLower.t[-1] + dt)
    return pendLower

# Purpose: Run Bashforth-Adams for some number of iterations on a driven
# pendulum
# Arguments: The pendulum to run B-A on
# Returns: The pendulum after being run for whatever the number of
# iterations
def runBashforthDriven(pendUpper, pendLower):
    for i in range(1, iterations):
        pendUpper = bashforthAdamsUpper(pendUpper,pendLower)
        pendLower = bashforthAdamsLower(pendUpper,pendLower)
    #return pend
```

```python
k = 1
gamma = .01
omega = .01

iterations = 1000
dt = .01

pend1 = Pendulum(1,0)
pend2 = Pendulum(1,0)

runBashforthDriven(pend1,pend2)
```

```python
plt.clf()
plt.plot(pend1.theta, pend1.vel)
plt.show()

plt.clf()
plt.plot(pend2.theta, pend2.vel)
plt.show()
```
